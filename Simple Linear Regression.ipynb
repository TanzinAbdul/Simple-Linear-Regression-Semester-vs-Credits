# Simple Linear Regression Project

This project demonstrates a basic implementation of linear regression using a simple dataset where the input variable is the **number of semesters completed**, and the output is the **number of credits earned**. 

Designed for beginners in machine learning, this project guides users through the process of:
- Import required libraries
- Load Dataset
- Data visualization
- Model training
- Predicting

By following this project, users will gain a practical understanding of regression analysis in an accessible and straightforward manner.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Load the data
CSV_PATH = '/Users/tanzin/Desktop/Research Interest (Responses) - Form Responses 1.csv' #paste your CSV file directory here
df = pd.read_csv(CSV_PATH)

# Assuming X is the feature(s) and Y is the target
X = df['Semester']  # Replace with your feature columns
Y = df['Credits Completed']  # Replace with your target column

# Step 2: Visualize the data
# If there's only one feature
plt.scatter(X, Y)
plt.xlabel('Semester')
plt.ylabel('Credits Completed')
plt.title('Semester vs credits completed')
plt.show()

# If there's more than one feature, you might need pair plots or subplots:
```

```python
# Step 3: Reshape X if there's only one feature

X = np.array(X)
X = X.reshape(-1, 1)   # Reshaping for a single feature

# Step 4: load the model
model = LinearRegression()

# model.fit(X_train, Y_train)
model.fit(X, Y)
```

```python
# Step 4: Predict using the same data (since you're not splitting)
Y_pred = model.predict(X)

# Step 5: Visualize the fitted line
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred, color='red', label='Fitted line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
# Step 5: Predict on new data
import numpy as np
new_x  = int(input('Enter your semester number: '))
new_data = np.array([[new_x]])  # Replace with new data points
new_data.reshape(-1,1)
predicted_value = model.predict(new_data)
print(f'Predicted value: {int(predicted_value[0])}')
```
